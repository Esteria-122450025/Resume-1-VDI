```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset dari file CSV
file_path = "/content/Spotify.csv"
spotify_data = pd.read_csv(file_path)
```

```python
# 1. Scatterplot: Popularity vs Danceability and Energy
plt.figure(figsize=(10, 6))
sns.scatterplot(x='danceability', y='popularity', hue='energy', size='energy', data=spotify_data, palette="cool", sizes=(20, 200))
plt.title("Popularity vs Danceability with Energy as Hue")
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.show()
```

### Penjelasan dan Analisis Scatter Plot

Scatter plot yang ditampilkan menggambarkan hubungan antara dua variabel: **danceability** (sumbu X) dan **popularity** (sumbu Y), dengan ukuran dan warna titik yang menunjukkan nilai **energy** dari setiap lagu.

#### 1. **Analisis Visual:**
- **Pola Umum**:
  - Terdapat distribusi yang cukup padat di bagian bawah sumbu Y (popularity), menunjukkan bahwa banyak lagu memiliki tingkat popularitas rendah (di bawah 40).
  - Sebagian besar lagu yang memiliki **danceability** antara 0.2 hingga 0.8 memiliki popularitas yang bervariasi, dengan beberapa lagu mencapai tingkat popularitas yang lebih tinggi.

- **Warna dan Ukuran Titik**:
  - Titik yang berwarna biru muda mewakili lagu-lagu dengan nilai **energy** rendah (0.0 hingga 0.4), sedangkan titik berwarna ungu mewakili lagu-lagu dengan **energy** tinggi (0.8 hingga 1.0).
  - Titik yang lebih besar menunjukkan nilai **energy** yang lebih tinggi. Ada tren di mana lagu-lagu yang lebih energik cenderung lebih populer.

#### 2. **Korelasi**:
- **Danceability vs Popularity**:
  - Terdapat indikasi bahwa **danceability** tidak sepenuhnya menentukan popularitas. Meskipun banyak lagu dengan **danceability** yang tinggi, popularitasnya bisa bervariasi.
  - Lagu dengan **danceability** yang lebih rendah tetapi dengan **energy** tinggi (ditunjukkan oleh warna ungu) juga dapat memiliki popularitas tinggi.

- **Energy**:
  - Dari visualisasi ini, tampaknya ada korelasi positif antara **energy** dan **popularity**. Lagu-lagu dengan **energy** yang lebih tinggi (berwarna ungu) lebih cenderung memiliki popularitas yang lebih tinggi dibandingkan dengan lagu-lagu berenergi rendah.
  
#### 3. **Insight**:
- **Target Pasar**:
  - Arti dari visualisasi ini sangat berguna bagi artis, produser, atau label musik untuk memahami elemen-elemen yang bisa meningkatkan popularitas sebuah lagu.
  - Artis yang ingin menciptakan lagu yang lebih populer mungkin harus mempertimbangkan untuk menciptakan musik yang lebih energik, terutama jika lagu tersebut juga memiliki tingkat **danceability** yang baik.

- **Strategi Pemasaran**:
  - Pemasar musik dapat menggunakan informasi ini untuk mempromosikan lagu-lagu dengan **energy** yang tinggi dan **danceability** yang baik, meningkatkan peluang untuk menjangkau audiens yang lebih luas.

```python
# 2. Boxplot: Distribution of Popularity by Track Genre
plt.figure(figsize=(12, 6))
sns.boxplot(x='track_genre', y='popularity', data=spotify_data)
plt.xticks(rotation=90)
plt.title("Distribution of Popularity by Track Genre")
plt.xlabel("Track Genre")
plt.ylabel("Popularity")
plt.show()
```

### Penjelasan dan Analisis Box Plot

Box plot yang ditampilkan menunjukkan distribusi popularitas lagu berdasarkan genre. Berikut adalah beberapa poin penting untuk analisis visualisasi ini:

#### 1. **Deskripsi Visual:**
- **Sumbu Y**: Mewakili **popularity** dengan rentang dari 0 hingga 100.
- **Sumbu X**: Menampilkan berbagai **track genre**. Nama-nama genre disusun di sepanjang sumbu X, meskipun sangat padat dan mungkin sulit dibaca.
- **Box Plot**: Setiap kotak (box) menunjukkan kuartil dari data, dengan garis di tengah kotak mewakili median popularitas. Garis vertikal (whiskers) menunjukkan rentang data dari minimum hingga maksimum, dan titik di luar whiskers adalah outliers (nilai ekstrem).

#### 2. **Analisis Data:**
- **Rentang Popularitas**:
  - Beberapa genre, seperti **hip hop** dan **rock**, tampaknya memiliki rentang popularitas yang lebih tinggi, dengan banyak nilai yang mendekati maksimum 100.
  - Genre lainnya seperti **classical** dan **jazz** menunjukkan rentang popularitas yang lebih rendah, yang dapat diindikasikan oleh median yang lebih rendah.

- **Outliers**:
  - Terdapat banyak outliers pada beberapa genre. Misalnya, **pop** dan **hip hop** memiliki banyak titik yang tersebar di atas nilai 80, menunjukkan bahwa ada beberapa lagu yang sangat populer dalam genre tersebut.
  - Outliers ini memberikan informasi tambahan tentang lagu-lagu yang mungkin sangat populer atau tidak biasa di luar distribusi normal untuk genre tersebut.

- **Variabilitas**:
  - **Rock**, **pop**, dan **hip hop** menunjukkan variabilitas yang tinggi dalam popularitas, yang terlihat dari whiskers yang panjang. Ini berarti bahwa di dalam genre ini, ada banyak lagu dengan tingkat popularitas yang berbeda-beda.
  - Beberapa genre seperti **ambient** dan **world music** mungkin menunjukkan variabilitas yang lebih rendah, dengan rentang interkuartil yang lebih sempit.

#### 3. **Insight**:
- **Tren Genre**:
  - Genre dengan popularitas yang lebih tinggi seperti **pop** dan **hip hop** menunjukkan bahwa lagu-lagu dalam genre ini lebih mungkin untuk menjadi hits dibandingkan genre lainnya.
  - Genre yang kurang populer seperti **classical** mungkin memiliki audiens yang lebih kecil, tetapi lagu-lagu dalam genre tersebut mungkin memiliki nilai seni atau nilai nostalgi yang tinggi.

- **Strategi Pemasaran**:
  - Bagi produser atau label musik, visualisasi ini membantu memahami di mana peluang berada. Genre yang menunjukkan potensi pertumbuhan dalam popularitas dapat menjadi target yang baik untuk promosi.

```python
# 3. Histogram: Distribution of Tempo and Valence
plt.figure(figsize=(10, 6))
sns.histplot(spotify_data['tempo'], kde=True, color='skyblue', label='Tempo')
sns.histplot(spotify_data['valence'], kde=True, color='orange', label='Valence')
plt.title("Distribution of Tempo and Valence")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

### Penjelasan dan Analisis Histogram

Histogram yang ditampilkan menggambarkan distribusi dua variabel: **tempo** dan **valence** dari lagu-lagu dalam dataset. Berikut adalah analisis dari visualisasi ini:

#### 1. **Deskripsi Visual:**
- **Sumbu Y**: Mewakili **frekuensi** atau jumlah lagu yang memiliki nilai tertentu untuk **tempo** dan **valence**.
- **Sumbu X**: Mewakili nilai-nilai dari **tempo** (dalam BPM) dan **valence** (yang mencerminkan positifitas atau emosi lagu).
- **Warna**:
  - **Biru**: Mewakili distribusi **tempo**.
  - **Kuning**: Mewakili distribusi **valence**.

#### 2. **Analisis Data:**
- **Distribusi Tempo**:
  - Distribusi tempo menunjukkan bahwa sebagian besar lagu memiliki tempo di antara 80 hingga 140 BPM, dengan puncak frekuensi pada sekitar 120 BPM. Ini mencerminkan bahwa banyak lagu pop dan genre lainnya cenderung memiliki tempo di kisaran tersebut.
  - Ada sedikit outliers di sisi kanan, yang menunjukkan bahwa ada beberapa lagu dengan tempo yang sangat tinggi, mungkin lagu-lagu EDM atau genre cepat lainnya.

- **Distribusi Valence**:
  - Distribusi valence menunjukkan bahwa banyak lagu memiliki nilai valence tinggi, dengan puncak frekuensi di antara 0.6 hingga 0.8. Ini menunjukkan bahwa banyak lagu dalam dataset ini memiliki karakteristik positif atau ceria.
  - Terdapat juga beberapa outliers pada sisi kiri (nilai rendah), menunjukkan bahwa ada lagu-lagu dengan emosi yang lebih gelap atau sedih.

#### 3. **Insight**:
- **Tren dalam Musik**:
  - Kombinasi dari kedua distribusi ini menunjukkan bahwa banyak lagu yang populer memiliki tempo yang moderat dan nilai valence yang tinggi, yang bisa jadi mengindikasikan preferensi pendengar terhadap musik yang ceria dan energik.
  - Lagu-lagu dengan tempo sangat tinggi dan valence yang rendah bisa jadi tidak terlalu umum, memberikan peluang bagi artis untuk bereksperimen dengan kombinasi yang berbeda.

- **Pemasaran dan Produksi**:
  - Bagi produser atau label musik, insight ini dapat digunakan untuk memproduksi lagu-lagu yang lebih mungkin diterima dengan baik oleh audiens. Memfokuskan pada nilai tempo dan valence yang lebih tinggi bisa jadi strategi yang baik untuk menciptakan lagu-lagu yang lebih menarik.

```python
df = pd.read_csv('/content/Spotify.csv')

# Menghitung korelasi antar variabel numerik
correlation_matrix = df[['danceability', 'energy', 'loudness', 'valence', 'tempo']].corr()
```

```python
# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Heatmap of Spotify Tracks Features')
plt.show()
```

#### Penjelasan dan Analisis Heatmap Korelasi Fitur Lagu Spotify

Heatmap korelasi adalah visualisasi yang menunjukkan hubungan antara berbagai variabel dalam sebuah dataset. Warna pada heatmap mewakili kekuatan dan arah korelasi:

* Warna merah: Korelasi positif yang kuat (semakin merah, semakin kuat hubungan positifnya)
* Warna biru: Korelasi negatif yang kuat (semakin biru, semakin kuat hubungan negatifnya)
* Warna putih: Tidak ada korelasi atau korelasi yang sangat lemah
Interpretasi Heatmap Fitur Lagu Spotify

**Kesimpulan dari hubungan antara berbagai fitur lagu di Spotify:**

* Energi dan Loudness: Terdapat korelasi positif yang sangat kuat antara fitur energi dan loudness. Ini berarti lagu-lagu dengan energi yang tinggi cenderung memiliki tingkat loudness yang tinggi pula. Lagu-lagu yang berenergi dan berisik cenderung dikelompokkan bersama.

* Danceability dan Valence: Ada korelasi positif yang moderat antara danceability dan valence. Ini menunjukkan bahwa lagu-lagu yang mudah untuk di-dance cenderung memiliki nada yang lebih positif atau menyenangkan.

* Energi, Loudness, dan Valence: Ketiga fitur ini memiliki korelasi positif satu sama lain. Ini menunjukkan bahwa lagu-lagu yang berenergi tinggi, berisik, dan memiliki nada positif cenderung dikelompokkan bersama.

* Tempo: Fitur tempo tidak menunjukkan korelasi yang kuat dengan fitur lainnya. Ini berarti kecepatan tempo lagu tidak terlalu terkait dengan fitur-fitur seperti energi, loudness, danceability, atau valence.
